# **Displaying and loading Images**

```

# Display one image
def display(a, title1 = "Original"):
    plt.imshow(a), plt.title(title1)
    plt.show()

```

```
# Loading
def get_image():
    
    # Reading 3 images to work
    img = cv2.imread("cat.png", cv2.IMREAD_UNCHANGED)[:,:,0]/256.0-0.5 
    return img
    

img=get_image()
display(img)    
```

# Main Function the heart of the program

```
def convolve(img,kernel):
  output=np.empty((img.shape[0]-2,img.shape[1]-2))
  for i in range(1,img.shape[0]-1):
    for j in range(1,img.shape[1]-1):
      sum=0
      for shiftI in range(-1,2):
        for shiftJ in range(-1,2):
          sum+=kernel[shiftI+1, shiftJ+1]*img[i+shiftI,j+shiftJ]
      output[i-1,j-1]=sum
  return output


  
kernel=np.array([[-0.5,1,-0.5],
                 [-0.5,1,-0.5],
                 [-0.5,1,-0.5]])

#kernel=np.array([[1,1,2],
#                 [1,1,2],
 #                [2,1,1]])
 
display(kernel)
result=convolve(img,kernel)
display(result,"result")    
```
